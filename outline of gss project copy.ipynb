## Data

The data used in this analysis is sourced from the General Social Survey (GSS), a survey that has been conducted since 1972 to capture general American socioeconomic opinions. The uniqueness of this survey lies in its longitudinal design, as it follows the same individuals over time, providing insights into changes at the individual level. However, the limitation is that only around 3000 people are surveyed within a year, and the dataset had notable missing values. It's crucial to note that questions regarding health and healthcare were introduced only in the year 2022. This is the data variables that our analysis focus on. This data, particularly the responses/insights into healthcare-related questions, holds significance for policymakers and government officials seeking to understand public sentiment on healthcare funding and policy. The observations suggest a potential disconnect between public needs and government implementation, especially if the data indicates widespread dissatisfaction or lack of confidence in the U.S. healthcare system. The delayed inclusion of healthcare questions raises questions about the prioritization of public opinion on such a vital topic within the realm of government decision-making.


Our project looks at five variables corresponding to health and healthcare: "HLTHINF," "HLTHGOV," "HLTHTAX," "HLTHSAT," and "CONHLTH". Reserved codes are also accounted for in these variables to account for missing data. The codes were ‘D’ for don’t know: this reserve code is used for when respondents do not know an answer to the question, ‘I’ for inapplicable: this reserve code is used when the respondent does not see a question because of structural factors or personal factors, ‘N’ for no answer, and ‘S’ for skipped on the web.

We faced a few challenges in the process of data cleaning. By checking for null values in the data frame of our chosen variables, there were 356,355 missing values across all cells. Furthermore, after more data cleaning to get rid of NA values, we discovered that 2022 was only year that the healthcare questions we chose were asked, thus limiting our analysis to one year instead of over time as we initially intended. Additionally, we also had to code the data as numeric according to the codebook tables and fill NAs/missings with '0'. After doing so, we had to filter our data to limit the observations to only meaningful ones (i.e. not '0'), which brought our total observations down from 72,392 to 1,008. This means we were working with a far smaller set of data than we originally believed due to a majority of the responses to our research variables were inapplicable ('I'). 

### Relevant Variables

**HLTHINF**: The variable “HLTHINF” is a numeric type that gauges respondents’ agreement or disagreement with a statement about the efficiency of the healthcare system in the United States. The statement reads: In general, the healthcare system in the United States is inefficient. The responses were categorized into five options: "STRONGLY AGREE," "AGREE," "NEITHER AGREE NOR DISAGREE," "DISAGREE," and "STRONGLY DISAGREE." For every option, there is a value associated with the corresponding response, frequencies, its unweighted percentage of individual responses, and an unweighted percentage excluding reserve codes. The summary reveals that a notable portion of respondents expressed concerns about the inefficiency of the health care system. Specifically, of the 1120 responses to the question, 37.1% of the respondents agreed and 22.9% of the respondents strongly agreed with the statement of the healthcare system being inefficient. (codebook pg 560). 


**HLTHGOV**: The variable “HLTHGOV” is a numeric measure capturing respondents’ opinions on the extent to which the government should provide healthcare services. Respondents were asked to respond with their opinion to this statement: “The government should provide only limited healthcare services.” The responses are categorized into five levels of agreement: "STRONGLY AGREE," "AGREE," "NEITHER AGREE NOR DISAGREE," "DISAGREE," and "STRONGLY DISAGREE." The summary table provides counts and percentages for each response category. Notably, of the 1096 responses, 417 respondents disagreed with the sentiment that the government should provide only limited healthcare services. Reserved codes are also accounted for in this summary and “NOT APPLICABLE” compromises the majority at 67.1%, suggesting a substantial portion of respondents might find the question not directly applicable to their circumstances. In conclusion, the table sheds light on varying degrees of agreement among respondents regarding the government’s role in providing healthcare services, offering a nuanced perspective on the diversity of opinions within the surveyed population.


**HLTHTAX**: The numeric variable “HLTHTAX” assesses respondents’ willingness to pay higher taxes to enhance the level of healthcare for all individuals in the United States. The responses are categorized into five levels of willingness: “VERY WILLING,” “FAIRLY WILLING,” “NEITHER WILLING NOR UNWILLING,” “FAIRLY UNWILLING,” and “VERY UNWILLING.” The majority of respondents express some degree of willingness to pay higher taxes to enhance healthcare quality. 2435 respondents did not answer the question, mainly because it was not applicable to them. Out of the 1109 respondents, the largest portion of people (345) voted for “FAIRLY WILLING” (31.1%). However, the next most popular options were “NEITHER WILLING NOR UNWILLING” (20.7%) and “VERY UNWILLING” (20.6%). This shows a very split perspective amongst participants. The positive response rates indicate a willingness among a considerable portion of the population to contribute more for enhanced health care services. However, the reservations expressed by another substantial segment highlight the need for nuanced policies that address varying perspectives on taxation for health care. 

**HLTHSAT**: The variable “HLTHSAT” measures respondents’ satisfaction with the healthcare system in the United States. The distribution of satisfaction levels reveals that “FAIRLY SATISFIED” (37.5%) is the predominant sentiment, followed by “FAIRLY DISSATISFIED” (16.7%). Many participants were also not applicable to this question, with 67.2% of total survey respondents not answering because the question was not applicable for them. There is a prominent satisfied population of respondents as 56% of answers were to some degree of satisfaction, but also a substantial portion were dissatisfied to some extent (28.9%) and around 15% of respondents had a neutral opinion.

**CONHLTH**: The last variable we are interested in, “CONHLTH”, assesses respondents’ confidence in the healthcare system by asking the question, “in general, how much confidence do you have in the healthcare system in the United States? The responses were categorized in five levels of confidence: “COMPLETE CONFIDENCE,” “A GREAT DEAL OF CONFIDENCE,” “SOME CONFIDENCE,” “VERY LITTLE CONFIDENCE,” AND “NO CONFIDENCE AT ALL.” A significant portion of respondents, totaling 32.4% of all answers, responded to the question. The other 67.6% of respondents did not respond mainly because the question was not applicable. The most significant percentage of real responses to the question were for the “SOME CONFIDENCE” answer with 47.4% of the valid responses being this answer.

## Visualizations

A statistical summary description table, a kernel density plot, and a bubble plot were generated for each independent-dependent variable pairing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
import warnings
# silencing warnings so cell outputs are not affected by personal OS differences
warnings.filterwarnings('ignore')
# setting size of graphs
# plt.rcParams['figure.dpi'] = 75
# we start by directly reading from the cleaned csv generated from the data cleaning performed in gss_data_cleaning.ipynb
total_df = pd.read_csv("./cleaned_data.csv")
```

**HLTHGOV VS CONHLTH**

```python
ctab = pd.crosstab(total_df['hlthgov'], total_df['conhlth'])
display(total_df.loc[:,['hlthgov', 'conhlth']].groupby('hlthgov').describe())
sns.kdeplot(data=total_df, x='conhlth', hue='hlthgov')
plt.show()
ctab = ctab.unstack().reset_index()
ctab.columns = list('XYS')
ctab['S'] = ctab['S'] * 10
plt.scatter(x='X', y='Y', s='S', data=ctab)
plt.xlabel("conhlth")
plt.ylabel("hlthgov")
plt.margins(0.1)
plt.show()
```

### Association

Below are calculated association values and extra crosstabulation tables for each pairing. The association value is calculated with Cramer's V, which is a measure of the degree to which the two variables are related; this value can help quantify which variable is the best predictor for each category. Since the smallest dimention of each varialbe is 5, we have the following interpretation values: 
- $<$ 0.05 shows a very weak relationship
- 0.05-0.1 shows a weak relationship
- 0.1-0.15 shows a moderate relationship
- 0.15-0.25 shows a strong relationship
- $>$ 0.25 shows a very strong relationship

This is first followed by a default crosstabulation table, and then followed by a normalized crosstab table. The crosstabulation tables contain information that is visualized by the bubble plots in the results section.

```python
# CONHLTH VS HLTHGOV
ctab = pd.crosstab(total_df['hlthgov'], total_df['conhlth'])
print(f"association: {ss.contingency.association(ctab)}")
display(ctab)
display(pd.crosstab(total_df['hlthgov'], total_df['conhlth'], normalize='all'))
```

```python
# CONHLTH VS HLTHTAX
ctab = pd.crosstab(total_df['hlthtax'], total_df['conhlth'])
print(f"association: {ss.contingency.association(ctab)}")
display(ctab)
display(pd.crosstab(total_df['hlthtax'], total_df['conhlth'], normalize='all'))
```
