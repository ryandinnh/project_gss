## Summary
This project investigates the relationship between government involvement, public satisfaction, and confidence levels in the United States healthcare system by analyzing data from the General Social Survey. The research question we intend to explore is: "How does the government's role impact satisfaction and confidence in the U.S. healthcare system?"
Examining data from 2022, when healthcare-related questions were introduced to the survey, reveals insights into respondents' perceptions during the pivotal time following the COVID-19 pandemic. Despite challenges like missing values and a limited dataset of 1,008 observations, the study provides a snapshot of public sentiment.
The analysis focuses on five variables: "HLTHINF" gauging healthcare system efficiency; "HLTHGOV" measuring opinions on government-provided healthcare services; "HLTHTAX" assessing willingness to pay higher taxes for improved healthcare; "HLTHSAT" measuring satisfaction levels; and "CONHLTH" evaluating confidence in the healthcare system.
Concerns about healthcare system inefficiency are highlighted, with a notable percentage expressing agreement. Opinions on the government's role vary, emphasizing the complexity of public perspectives. The willingness to pay higher taxes for healthcare exhibits a split perspective, underlining the need for nuanced policies.
Satisfaction with the healthcare system indicates a predominant fairly satisfied sentiment, while confidence levels show a considerable proportion with only some confidence. The dataset also reveals dissatisfaction and a neutral opinion.
Visualizations, including kernel density plots and bubble plots, accompany cross-tabulations to elucidate our results. The findings contribute valuable information for policymakers and government entities who might be starting to value public need and opinion more regarding healthcare, considering the addition of these questions to the survey. In conclusion, this research offers a comprehensive exploration of public opinions on healthcare, highlighting the need for ongoing studies to capture evolving sentiments and inform responsive policy decisions.

## Data

The data used in this analysis is sourced from the General Social Survey (GSS), a survey that has been conducted since 1972 to capture general American socioeconomic opinions. The uniqueness of this survey lies in its longitudinal design, as it follows the same individuals over time, providing insights into changes at the individual level. However, the limitation is that only around 3000 people are surveyed within a year, and the dataset had notable missing values. It's crucial to note that questions regarding health and healthcare were introduced only in the year 2022. This is the data variables that our analysis focus on. Our project looks at five variables corresponding to health and healthcare: "HLTHINF," "HLTHGOV," "HLTHTAX," "HLTHSAT," and "CONHLTH". Reserved codes are also accounted for in these variables to account for missing data. The codes were ‘D’ for don’t know: this reserve code is used for when respondents do not know an answer to the question, ‘I’ for inapplicable: this reserve code is used when the respondent does not see a question because of structural factors or personal factors, ‘N’ for no answer, and ‘S’ for skipped on the web.

We faced a few challenges in the process of data cleaning, which can be seen in the `gss_data_cleaning.ipynb` file. By checking for null values in the data frame of our chosen variables, there were 356,355 missing values across all cells. Furthermore, after more data cleaning to get rid of NA values, we discovered that 2022 was only year that the healthcare questions we chose were asked, thus limiting our analysis to one year instead of over time as we initially intended. Additionally, we also had to code the data as numeric according to the codebook tables and fill NAs/missings with '0'. After doing so, we had to filter our data to limit the observations to only meaningful ones (i.e. not '0'), which brought our total observations down from 72,392 to 1,008. This means we were working with a far smaller set of data than we originally believed due to a majority of the responses to our research variables were inapplicable ('I').

### Relevant Variables

**HLTHINF**: The variable “HLTHINF” is a numeric type that gauges respondents’ agreement or disagreement with a statement about the efficiency of the healthcare system in the United States. The statement reads: In general, the healthcare system in the United States is inefficient. The responses were categorized into five options: "STRONGLY AGREE," "AGREE," "NEITHER AGREE NOR DISAGREE," "DISAGREE," and "STRONGLY DISAGREE." For every option, there is a value associated with the corresponding response, frequencies, its unweighted percentage of individual responses, and an unweighted percentage excluding reserve codes. The summary reveals that a notable portion of respondents expressed concerns about the inefficiency of the health care system. Specifically, of the 1120 responses to the question, 37.1% of the respondents agreed and 22.9% of the respondents strongly agreed with the statement of the healthcare system being inefficient. (codebook pg 560).


**HLTHGOV**: The variable “HLTHGOV” is a numeric measure capturing respondents’ opinions on the extent to which the government should provide healthcare services. Respondents were asked to respond with their opinion to this statement: “The government should provide only limited healthcare services.” The responses are categorized into five levels of agreement: "STRONGLY AGREE," "AGREE," "NEITHER AGREE NOR DISAGREE," "DISAGREE," and "STRONGLY DISAGREE." The summary table provides counts and percentages for each response category. Notably, of the 1096 responses, 417 respondents disagreed with the sentiment that the government should provide only limited healthcare services. Reserved codes are also accounted for in this summary and “NOT APPLICABLE” compromises the majority at 67.1%, suggesting a substantial portion of respondents might find the question not directly applicable to their circumstances. In conclusion, the table sheds light on varying degrees of agreement among respondents regarding the government’s role in providing healthcare services, offering a nuanced perspective on the diversity of opinions within the surveyed population.


**HLTHTAX**: The numeric variable “HLTHTAX” assesses respondents’ willingness to pay higher taxes to enhance the level of healthcare for all individuals in the United States. The responses are categorized into five levels of willingness: “VERY WILLING,” “FAIRLY WILLING,” “NEITHER WILLING NOR UNWILLING,” “FAIRLY UNWILLING,” and “VERY UNWILLING.” The majority of respondents express some degree of willingness to pay higher taxes to enhance healthcare quality. 2435 respondents did not answer the question, mainly because it was not applicable to them. Out of the 1109 respondents, the largest portion of people (345) voted for “FAIRLY WILLING” (31.1%). However, the next most popular options were “NEITHER WILLING NOR UNWILLING” (20.7%) and “VERY UNWILLING” (20.6%). This shows a very split perspective amongst participants. The positive response rates indicate a willingness among a considerable portion of the population to contribute more for enhanced health care services. However, the reservations expressed by another substantial segment highlight the need for nuanced policies that address varying perspectives on taxation for health care.

**HLTHSAT**: The variable “HLTHSAT” measures respondents’ satisfaction with the healthcare system in the United States. The distribution of satisfaction levels reveals that “FAIRLY SATISFIED” (37.5%) is the predominant sentiment, followed by “FAIRLY DISSATISFIED” (16.7%). Many participants were also not applicable to this question, with 67.2% of total survey respondents not answering because the question was not applicable for them. There is a prominent satisfied population of respondents as 56% of answers were to some degree of satisfaction, but also a substantial portion were dissatisfied to some extent (28.9%) and around 15% of respondents had a neutral opinion.

**CONHLTH**: The last variable we are interested in, “CONHLTH”, assesses respondents’ confidence in the healthcare system by asking the question, “in general, how much confidence do you have in the healthcare system in the United States? The responses were categorized in five levels of confidence: “COMPLETE CONFIDENCE,” “A GREAT DEAL OF CONFIDENCE,” “SOME CONFIDENCE,” “VERY LITTLE CONFIDENCE,” AND “NO CONFIDENCE AT ALL.” A significant portion of respondents, totaling 32.4% of all answers, responded to the question. The other 67.6% of respondents did not respond mainly because the question was not applicable. The most significant percentage of real responses to the question were for the “SOME CONFIDENCE” answer with 47.4% of the valid responses being this answer.


## Visualizations and Analysis

A statistical summary description table, a kernel density plot, and a bubble plot were generated for each independent-dependent variable pairing. Other than numbers refering to how the variables were coded, specific values mentioned in the written analysis can be found in the appendix tables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
# silencing warnings so cell outputs are not affected by personal OS differences
import warnings
warnings.filterwarnings('ignore')
# we start by directly reading from the cleaned csv generated from the data cleaning performed in gss_data_cleaning.ipynb
df = pd.read_csv("./gss_selected_clean.csv")
```

**HLTHGOV VS CONHLTH**

```python
ctab = pd.crosstab(df['hlthgov'], df['conhlth'])
display(df.loc[:,['hlthgov', 'conhlth']].groupby('hlthgov').describe())
sns.kdeplot(data=df, x='conhlth', hue='hlthgov')
plt.show()
ctab = ctab.unstack().reset_index()
ctab.columns = list('XYS')
ctab['S'] = ctab['S'] * 10
plt.scatter(x='X', y='Y', s='S', data=ctab)
plt.xlabel('conhlth')
plt.ylabel('hlthgov')
plt.margins(0.1)
plt.show()
```

For the variables ‘HLTHGOV’ and ‘HLTHTAX’, the tables and visualizations led to the finding: As the belief the government should provide only limited health care services decreases, the confidence that people have in the healthcare system in the United States also decreases. This trend continues until people have some confidence in the health care system in the United States and disagree that the government should provide only limited health care services. This is evident by the bubbles continuously getting larger until point (conhlth: 3, hlthgov: 4) and then progressively getting smaller past this point. This finding is also supported by the normalized cross tabulation of variables “HLTHGOV” and “CONHLTH”, that can be found in the appendix, where point (3.0, 4) has the highest value of 0.183. The concentration of larger points and values could be due to the fact that in 2022 Covid case numbers rose from 2021, but deaths decreased which could explain how a majority of respondents still had some confidence in the US Healthcare system and also thought that more government aid should be supplied, as it was reducing overall covid19 death. To explain the decreasing trend between belief and confidence, when people believe that the government should play a more active role in healthcare, they may also likely expect higher standards of quality and affordability. If these expectations are not met, it can lead to decreased confidence in the healthcare system and vice versa.

### Data limitations
#### Missing data
In order to conduct analysis with valid data, we first manipulate the missing data, which removes 356355 observations. This can be a big portion of the whole data set, and these missing data affect the completeness and representativeness of the dataset, influencing the result accuracy/precision of our analysis.
##### Non-response bias
This is from the null values we cleaned, people prefer not to answer some specific type of questions. Non-response bias affects the data result seriously, reducing the confidence in the validity and generalizability of study findings. Sometimes, it is also linked to survivorship bias.
#### Sample size
The codebook talks about oversampling, which means a noticeable portion of the sample is weighted a certain way for the study.
Additionally, after doing data cleaning, we have a sample size of 1008, which can be a small sample size compared to the whole population in the U.S. When conducting sociological research, managing a representative sample is crucial for making valid analysis. A sample size of 1008 can not properly represent the diversity and complexity of the population under our study. This limits the generalizability or credibility of our findings.
The lack of statistical power also limits the ability of our study to conclude or detect important and clear relationships between variables.
#### Lack of data with more time series
After data cleaning, it’s surprising that all useful data left is from 2022. Meanwhile, having data for only one year, 2022, limits the ability of us to observe long-term trends over time. Since we are conducting social research, sociological phenomena tend to evolve through years, and data spanning several years may be needed to capture meaningful patterns in social dynamics.
Besides, as the data is confined to a single year, it may cause potential temporal bias, meaning that the results or analysis we make can only be established under this specific year 2022. It may reflect specific/certain circumstances that happened during 2022. There’s no doubt that there’s a noticable circumstance existing, which is COVID-19. COVID-19 has a strong relation with the healthcare industry, so the influence of COVID-19 may cause some specific results during 2022, but not in other years.
With only cross-sectional data available for the year 2022, it is very difficult to determine the causality between variables.
#### Confounding variables
When conducting the study research, we find out that the independent variables that we can use to solve the research question may not be enough to answer why certain relationships exist, as there may be confounding variables such as different demographic variables. Politicla affiliation may also play an important role as different parties tend to have different attitudes about government, so we may lack different types of variables.
### Research strategy evaluation
The strategy of this study/research need to be improved and developed due to several problems.
#### The wording of survey questions
Some of the questions are kind of leading. During surveys of data collection, leading questions can influence respondents' answers since the wording of the question can imply the answer the respondent wants or the answer the respondent tends to avoid.
#### The format of survey questions
The format design of the survey questions are the same, maybe we can use different types of research survey questions to incentify people’s interest in the survey.
#### The amount of variables
The number of independent variables is not enough to make a generalized conclusion. We need more possible independent variables that might have cause-and-effect relationships with satisfaction/confidence level.
#### The methods to “publicize”/ “spread” surveys
According to the value of the sample data, we think there’s a lack of sample data. It might be caused by the inefficient methods of data collection. We suppose that the researchers can add some specific motivation and conduct surveys through different platforms/methods to incentify people to respond to the survey.
#### Revision after receiving feedback
It’s proper for researchers to develop and revise the survey questions based on feedback in the early stages they received. However, there were too many null values that were meaningless in the final dataset. It might be necessary to take time to revise the questions.

### Possible improvements
Based on the problems and inadequacy we noticed, it’s better to develop and redesign survey questions.

Firstly, revise the wording/format of questions. For example, “How willing would you be to pay higher taxes to improve the level of health care for all people in the United States? Would you be…” This question is leading since “higher tax” is commonly unexpected. It could be changed into “What kind of change in tax level do you think that could improve the level of healthcare for all people in the U.S?--1. More tax 2. Stay as it was 3. Less tax 4.other____”. In this way, people may be more willing to respond and make more real answers from deep in their mind.

Secondly, it’s necessary to add more survey questions to test more possible independent variables that may have cause-and-effect relationships with our dependent variable. For instance, there are some demographic variables that may have influences such as age, gender, political party, etc. Besides, 2022 is a special year, since the entire globe is being influenced by COVID-19, so COVID-19 might be a key variable in this study. It’s proper to create a question that tests the relationship between COVID-19 and the healthcare system.

Thirdly, to encourage more people to participate, there are several different incentives types. Some usual types are money, service, gift cards, or some specific products that people will have interest in. By adding some incentives, people will be more willing to engage. Meanwhile, utilizing different platforms to publicize the survey like sharing in social media can expand our sample range. It’s better for researchers to use different methods of data collection when facing people with different age ranges. For the old, maybe it’s better to do the survey in person at places where the old prefer to go. For young people, using online surveys can be a good option.
